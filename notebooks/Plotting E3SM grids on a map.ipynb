## E3SM plot AODVIS on a map

```python
import xarray
import numpy as np
from matplotlib import pyplot as plt
from cartopy import crs as ccrs
```

```python
file_path = '/lcrc/group/earthscience/rjackson/e3sm_data/Dustv1_edison_1yr_ne120_dsd_extAN_mamAN_dstemi095_noforc.cam.h0.0001-12.climo.nc'
```

```python
ds = xarray.open_dataset(file_path)
```

```python
ds
```

```python
plt.figure(figsize=(10,5))
ax = plt.axes(projection = ccrs.PlateCarree())
ds.AODVIS.isel(time=0).plot.contourf(axis=ax, vmin=0, vmax=4, cmap='binary', levels=np.linspace(0, 4, 50))
ax.set_global(); ax.coastlines();
```
